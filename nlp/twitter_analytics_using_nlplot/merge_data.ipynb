```python
df1 = pd.read_csv('twitter_ai.csv')
df2 = pd.read_csv('twitter_ds.csv')
df3 = pd.read_csv('twitter_kaggle.csv')
df4 = pd.read_csv('twitter_python.csv')
```

```python
cols = ['searched_for', 'hashtags', 'hashtag_len']
df1 = df1[cols]
df2 = df2[cols]
df3 = df3[cols]
df4 = df4[cols]
```

```python
df = pd.concat([df1, df2], axis=0)
df = pd.concat([df, df3], axis=0)
df = pd.concat([df, df4], axis=0)
```
